## fillna()
this will fill the nan value with the value you pass

```python
#this will replace all the NaN values with 0 
zerodf=df.fillna(0)
```

```python
zerodf.isnull().any() 
# all the NaN values are now replaced with 0 
```

```python
# example on random null data point
print('before fillna()',df.iloc[300]['Sunshine'])
print('after fill na()',zerodf.iloc[300]['Sunshine'])
```

in this case for wind dir we cant write 0. 

lets assume we want to put wind direction SW in WindGustDir

then

```python
newdf=df.fillna({
    'Sunshine':0,
    'WindGustDir':'SW',
    'WindGustSpeed':0
    
})
```

```python
# now  all NaN values in WindGustDir will be replaced with 'SW'
print('before fillna()',df.iloc[193]['WindGustDir'])
print('after fill na()',newdf.iloc[193]['WindGustDir'])
```

```python
# we can also fill values with mean median in some cases 
# e.g.
meanfilldf=df.fillna(df.mean())
```


# fillna() methods

this will fill the NaN values with predefined methods

methods={‘backfill’, ‘bfill’, ‘pad’, ‘ffill’, None}

default None
Method to use for filling holes in reindexed Series pad / 

ffill: propagate last valid observation forward to next valid backfill /

bfill: use next valid observation to fill gap.

```python
# fill na will fill the Nan values with previous value of it
# new dataframe for easier understanding 
df=pd.read_csv('new_weather_data.csv',parse_dates=['day'])
```

```python
df
```

```python
df.fillna(method='ffill') # all the NaN values are now filled with previous one
```

```python
df.fillna(method='bfill') # it fills nan value with next value to it
```

```python
# we can also copy values horizontally
df.fillna(method='bfill',axis='columns')
```

```python
# we can also set limit it wont fill the null upto that limit 
df.fillna(method='ffill',limit=1)
```

# Concatination


we do concatination when we want to join two or more dataframes

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
# joining them together
pd.concat([india_weather,us_weather]) # returns joined df with their index
```

```python
# fixing index
pd.concat([india_weather,us_weather],ignore_index=True)#now we get continious index
```

```python
# creating a keys
cdf=pd.concat([india_weather,us_weather],keys=['india','us'])# creates an extra index
cdf
```

```python
# indexing in such df
cdf.loc['india']
```

```python
cdf.loc['us']
```

here we just appended the rows now we will do appending data frames as column

```python
# now guess
```

```python

temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
# now ideally we want to see windspeed as appended column  in original dat frame
# to do such column concat we do axis argument
```

```python
pd.concat([windspeed_df,temperature_df],axis=1) # here it is appending data frame as it is verticly 
# if order of cities are different in both dataframes then this will cause problems
#in such case we can add index argument in dataframe
```

# Merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
#merge is same as joins in datbase
pd.merge(df1,df2,on='city')
```

```python
# if you have extra city in one column
df2 = pd.DataFrame({
    "city": ["chicago","new york",],
    "humidity": [65,68],
})
df2
```

```python
pd.merge(df1,df2,on='city')# this will merge only those who are present in both
```


### joins
outer join,inner join,left,right


```python
pd.merge(df1,df2,on='city',how='outer') #outer join will put NaN value where values are not present
# and default is inner join as shown above
```
