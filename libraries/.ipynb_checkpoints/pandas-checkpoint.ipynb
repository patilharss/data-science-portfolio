## Conditional data selction from data frame


```python
# checking for MaxTemp column where value is greater than or equal to 35 
# only this will return us the true false value at that row depanding upon our condition
df['MaxTemp']>=35 
```

```python
# and if we want to get full data frame then we can do
df[df['MaxTemp']>=35] 
# so this will return us data frame with rows and columns where value of maxtemp is >= 35
```

```python
#getting a row where rainfall was max

df[df.Rainfall==df.Rainfall.max()]  # or df[df[Rainfall]==df[Rainfall].max()]


# this will return the df of max rainfall value
```



-----
if you want to print only specific columns
like if you want to only print minTemp and maxTemp when rainfall is lowest.
then



```python
df[['MaxTemp','MinTemp','Rainfall']][df['Rainfall']==df['Rainfall'].max()]
```

**set_index()**

df have default index starting from 0 


if you want to make something else a index then

```python
df.index #this will return info abt index
```

assume if you want to set column as your index then

***df.set_index(column_name)***

```python
df2=df.set_index('MinTemp') # or df.set_index('MinTemp',inplace=True) this will make change in original df
df.set_index('MinTemp') # This will return dataframe where index in mintemp
```

and now we can use **loc** on this df

iloc only work with int inde

where loc use with any dtype

```python
df2.loc[9.5] # so this will return us the value of rows where index i.e mintemp (int this case) is 9.5 
```

we can also reset the index using **reset_index**

```python
df2.reset_index(inplace=True)
```
